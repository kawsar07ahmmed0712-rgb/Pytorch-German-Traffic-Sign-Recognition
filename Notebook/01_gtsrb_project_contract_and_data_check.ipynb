## 3. Load dataset CSV files

The GTSRB dataset includes CSV files that contain important information about the images and labels.

Before loading images directly, we will inspect these CSV files to understand how the dataset is organized.

### Guiding questions

- What columns are available in `Train.csv`, `Test.csv`, and `Meta.csv`?
- How many rows are in each CSV file?
- Which column contains the image path?
- Which column contains the class label?

```python
import pandas as pd

train_csv_path = data_dir / "Train.csv"
test_csv_path = data_dir / "Test.csv"
meta_csv_path = data_dir / "Meta.csv"

train_df = pd.read_csv(train_csv_path)
test_df = pd.read_csv(test_csv_path)
meta_df = pd.read_csv(meta_csv_path)

print("Train CSV shape:", train_df.shape)
print("Test CSV shape:", test_df.shape)
print("Meta CSV shape:", meta_df.shape)
```

```python
print("Train CSV columns:")
print(train_df.columns.tolist())

print("\nTest CSV columns:")
print(test_df.columns.tolist())

print("\nMeta CSV columns:")
print(meta_df.columns.tolist())
```

### Observation

The dataset CSV files were loaded successfully.

- `Train.csv` has **39,209 rows** and **8 columns**
- `Test.csv` has **12,630 rows** and **8 columns**
- `Meta.csv` has **43 rows** and **5 columns**

Both `Train.csv` and `Test.csv` contain the same important columns:

- `Width`
- `Height`
- `Roi.X1`, `Roi.Y1`, `Roi.X2`, `Roi.Y2`
- `ClassId`
- `Path`

The most important columns for this project are:

- `Path`: image file path
- `ClassId`: target label/class

`Meta.csv` contains extra information about the 43 traffic sign classes.

### Decision

The CSV structure looks correct.  
Before ending this setup notebook, we will do one basic integrity check for missing values, duplicate rows, class count, and sample image paths.

## 4. Basic dataset integrity check

Before finishing this notebook, we will run a small sanity check on the CSV files.

### Guiding questions

- Are there any missing values?
- Are there duplicate rows?
- How many unique classes are available?
- Do the image paths look valid?

```python
print("Missing values in Train.csv:")
print(train_df.isna().sum())

print("\nMissing values in Test.csv:")
print(test_df.isna().sum())

print("\nDuplicate rows:")
print("Train duplicates:", train_df.duplicated().sum())
print("Test duplicates:", test_df.duplicated().sum())

print("\nUnique classes:")
print("Train unique classes:", train_df["ClassId"].nunique())
print("Test unique classes:", test_df["ClassId"].nunique())
print("Meta unique classes:", meta_df["ClassId"].nunique())

print("\nSample train image paths:")
print(train_df["Path"].head().tolist())

print("\nSample test image paths:")
print(test_df["Path"].head().tolist())
```

### Observation

The basic dataset integrity check looks clean.

- `Train.csv` has **no missing values**
- `Test.csv` has **no missing values**
- There are **no duplicate rows** in both train and test CSV files
- Train, test, and meta files all contain **43 unique classes**
- Image paths are available in the `Path` column
- Train image paths follow this structure: `Train/class_id/image_name.png`
- Test image paths follow this structure: `Test/image_name.png`

This means the CSV files are clean enough to continue with image-level exploration.
